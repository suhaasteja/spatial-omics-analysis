```python
import torch
import torch.nn as nn
import numpy as np
import pandas as pd
import plotly.express as px
from sklearn.preprocessing import StandardScaler
import anndata
from sklearn.neighbors import NearestNeighbors
import scanpy as sc
```

```python
class SimpleSpatialClassifier(nn.Module):
    def __init__(self, n_features, n_classes):
        super().__init__()
        # Combine gene expression and spatial features
        self.network = nn.Sequential(
            nn.Linear(n_features + 12, 128),  # 12 spatial features (2 coords + 10 neighbor features)
            nn.ReLU(),
            nn.Dropout(0.2),
            nn.Linear(128, 64),
            nn.ReLU(),
            nn.Dropout(0.2),
            nn.Linear(64, n_classes)
        )

    def forward(self, x):
        return torch.log_softmax(self.network(x), dim=1)

class SpatialAnalyzer:
    def __init__(self):
        self.model = None
        self.scaler = StandardScaler()

    def preprocess_data(self, adata):
        print("Preprocessing data...")
        # 1. Process gene expression
        sc.pp.normalize_total(adata)
        sc.pp.log1p(adata)

        # 2. Get spatial coordinates
        coords = adata.obsm['spatial']

        # 3. Find nearest neighbors
        n_neighbors = 5
        nbrs = NearestNeighbors(n_neighbors=n_neighbors).fit(coords)
        distances, indices = nbrs.kneighbors(coords)

        # 4. Create features that include spatial information
        # For each cell, we'll include:
        # - Its coordinates (x, y)
        # - Average expression of its 5 nearest neighbors (adds local context)
        neighbor_avg = np.zeros((len(coords), adata.X.shape[1]))
        for i in range(len(coords)):
            neighbor_avg[i] = adata.X[indices[i]].mean(axis=0)

        # 5. Combine all features
        gene_expression = self.scaler.fit_transform(adata.X)
        spatial_features = np.hstack([
            coords,                    # x, y coordinates
            neighbor_avg[:, :10]       # first 10 genes of neighbors (for simplicity)
        ])

        combined_features = np.hstack([gene_expression, spatial_features])
        return torch.FloatTensor(combined_features)

    def train(self, adata, labels, epochs=50):
        print("Preparing training data...")
        features = self.preprocess_data(adata)
        n_features = adata.X.shape[1]  # number of genes
        n_classes = len(np.unique(labels))

        # Initialize model
        self.model = SimpleSpatialClassifier(n_features, n_classes)
        optimizer = torch.optim.Adam(self.model.parameters(), lr=0.01)
        criterion = nn.NLLLoss()

        # Convert labels to tensor
        labels = torch.LongTensor(labels)

        print("Starting training...")
        from IPython.display import clear_output
        import matplotlib.pyplot as plt

        losses = []
        for epoch in range(epochs):
            self.model.train()
            optimizer.zero_grad()

            # Forward pass
            outputs = self.model(features)
            loss = criterion(outputs, labels)

            # Backward pass
            loss.backward()
            optimizer.step()

            losses.append(loss.item())

            # Display progress
            if (epoch + 1) % 5 == 0:
                clear_output(wait=True)
                plt.plot(losses)
                plt.title(f'Training Loss (Epoch {epoch+1})')
                plt.show()
                print(f'Epoch {epoch+1}/{epochs}, Loss: {loss.item():.4f}')

    def predict(self, adata):
        features = self.preprocess_data(adata)
        self.model.eval()
        with torch.no_grad():
            outputs = self.model(features)
        return torch.exp(outputs).numpy()

# Function to create sample data
def create_sample_data(n_cells=1000, n_genes=50):
    # Create synthetic expression matrix
    X = np.random.normal(size=(n_cells, n_genes))

    # Create synthetic spatial coordinates in a circle-like pattern
    theta = np.random.uniform(0, 2*np.pi, n_cells)
    r = np.random.normal(50, 10, n_cells)
    x = r * np.cos(theta)
    y = r * np.sin(theta)
    spatial_coords = np.column_stack([x, y])

    # Create synthetic cell types based on location
    # Cells in different regions will have different types
    distances = np.sqrt(x**2 + y**2)
    labels = np.zeros(n_cells)
    labels[distances < 30] = 0    # Inner circle
    labels[(distances >= 30) & (distances < 60)] = 1  # Middle ring
    labels[distances >= 60] = 2   # Outer ring

    # Create AnnData object
    adata = anndata.AnnData(X=X)
    adata.obsm['spatial'] = spatial_coords

    return adata, labels
```

```python
# Create and train on sample data
adata, labels = create_sample_data()
analyzer = SpatialAnalyzer()
analyzer.train(adata, labels, epochs=30)

# Visualize results
predictions = analyzer.predict(adata)
coords = adata.obsm['spatial']
df = pd.DataFrame({
    'x': coords[:, 0],
    'y': coords[:, 1],
    'true_type': labels,
    'predicted_type': predictions.argmax(axis=1)
})

# Plot true vs predicted
fig1 = px.scatter(df, x='x', y='y', color='true_type', title='True Cell Types')
fig2 = px.scatter(df, x='x', y='y', color='predicted_type', title='Predicted Cell Types')
fig1.show()
fig2.show()
```

```python
# Better synthetic data generation
def create_sample_data(n_cells=1000, n_genes=50):
    # Create more realistic gene expression patterns
    np.random.seed(42)  # For reproducibility

    # Generate cell positions in a more structured way
    theta = np.random.uniform(0, 2*np.pi, n_cells)
    r = np.random.normal(50, 10, n_cells)
    x = r * np.cos(theta)
    y = r * np.sin(theta)
    spatial_coords = np.column_stack([x, y])

    # Create distinct gene expression patterns for each cell type
    X = np.zeros((n_cells, n_genes))
    distances = np.sqrt(x**2 + y**2)

    # Assign cell types
    labels = np.zeros(n_cells)
    labels[distances < 30] = 0    # Inner circle
    labels[(distances >= 30) & (distances < 60)] = 1  # Middle ring
    labels[distances >= 60] = 2   # Outer ring

    # Generate distinct gene expression patterns for each cell type
    for i in range(n_cells):
        if labels[i] == 0:
            X[i] = np.random.normal(2, 0.5, n_genes)  # Type 0 pattern
        elif labels[i] == 1:
            X[i] = np.random.normal(5, 0.5, n_genes)  # Type 1 pattern
        else:
            X[i] = np.random.normal(8, 0.5, n_genes)  # Type 2 pattern

    # Add some noise
    X += np.random.normal(0, 0.1, (n_cells, n_genes))

    # Create AnnData object
    adata = anndata.AnnData(X=X)
    adata.obsm['spatial'] = spatial_coords

    return adata, labels

# Improved model architecture
class ImprovedSpatialClassifier(nn.Module):
    def __init__(self, n_features, n_classes):
        super().__init__()

        # Separate networks for gene expression and spatial features
        self.gene_network = nn.Sequential(
            nn.Linear(n_features, 128),
            nn.BatchNorm1d(128),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(128, 64)
        )

        self.spatial_network = nn.Sequential(
            nn.Linear(12, 32),  # 2 coords + 10 neighbor features
            nn.BatchNorm1d(32),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(32, 16)
        )

        # Combined network
        self.combined_network = nn.Sequential(
            nn.Linear(80, 64),  # 64 + 16 features
            nn.BatchNorm1d(64),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(64, n_classes)
        )

    def forward(self, x):
        # Split input into gene expression and spatial features
        gene_features = x[:, :-12]
        spatial_features = x[:, -12:]

        # Process separately
        gene_out = self.gene_network(gene_features)
        spatial_out = self.spatial_network(spatial_features)

        # Combine features
        combined = torch.cat([gene_out, spatial_out], dim=1)

        # Final processing
        out = self.combined_network(combined)
        return torch.log_softmax(out, dim=1)

# Improved training process
class ImprovedSpatialAnalyzer:
    def __init__(self):
        self.model = None
        self.scaler = StandardScaler()
        self.device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

    def preprocess_data(self, adata):
        print("Preprocessing data...")
        # Basic preprocessing
        sc.pp.normalize_total(adata)
        sc.pp.log1p(adata)

        # Get spatial coordinates
        coords = adata.obsm['spatial']

        # Find nearest neighbors
        n_neighbors = 5
        nbrs = NearestNeighbors(n_neighbors=n_neighbors).fit(coords)
        distances, indices = nbrs.kneighbors(coords)

        # Create neighbor features
        neighbor_avg = np.zeros((len(coords), adata.X.shape[1]))
        for i in range(len(coords)):
            neighbor_avg[i] = adata.X[indices[i]].mean(axis=0)

        # Scale features
        gene_expression = self.scaler.fit_transform(adata.X)
        spatial_features = np.hstack([
            coords,
            neighbor_avg[:, :10]
        ])

        # Combine features
        combined_features = np.hstack([gene_expression, spatial_features])
        return torch.FloatTensor(combined_features).to(self.device)

    def train(self, adata, labels, epochs=100):
        print("Preparing training data...")
        features = self.preprocess_data(adata)
        n_features = adata.X.shape[1]
        n_classes = len(np.unique(labels))

        # Initialize model
        self.model = ImprovedSpatialClassifier(n_features, n_classes).to(self.device)
        optimizer = torch.optim.Adam(self.model.parameters(), lr=0.001)
        scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(optimizer, mode='min', factor=0.5, patience=5)
        criterion = nn.NLLLoss()

        # Convert labels to tensor
        labels = torch.LongTensor(labels).to(self.device)

        print("Starting training...")
        best_loss = float('inf')
        early_stopping_counter = 0
        early_stopping_patience = 10

        losses = []
        for epoch in range(epochs):
            self.model.train()
            optimizer.zero_grad()

            outputs = self.model(features)
            loss = criterion(outputs, labels)

            loss.backward()
            optimizer.step()

            losses.append(loss.item())

            # Learning rate scheduling
            scheduler.step(loss)

            # Early stopping
            if loss < best_loss:
                best_loss = loss
                early_stopping_counter = 0
            else:
                early_stopping_counter += 1

            if early_stopping_counter >= early_stopping_patience:
                print(f"Early stopping at epoch {epoch}")
                break

            if (epoch + 1) % 10 == 0:
                print(f'Epoch {epoch+1}/{epochs}, Loss: {loss.item():.4f}')

    def predict(self, adata):
        features = self.preprocess_data(adata)
        self.model.eval()
        with torch.no_grad():
            outputs = self.model(features)
        return torch.exp(outputs).cpu().numpy()
```

```python
# Create and train on sample data
adata, labels = create_sample_data()
analyzer = ImprovedSpatialAnalyzer()
analyzer.train(adata, labels, epochs=100)

# Visualize results
predictions = analyzer.predict(adata)
coords = adata.obsm['spatial']
df = pd.DataFrame({
    'x': coords[:, 0],
    'y': coords[:, 1],
    'true_type': labels,
    'predicted_type': predictions.argmax(axis=1)
})

# Plot true vs predicted
fig1 = px.scatter(df, x='x', y='y', color='true_type', title='True Cell Types')
fig2 = px.scatter(df, x='x', y='y', color='predicted_type', title='Predicted Cell Types')
fig1.show()
fig2.show()
```
